```python
import plotly.express as px
import plotly.io as pio
pio.renderers.default = "iframe_connected"
```

```python
world = df[df['iso_code'].str.len() == 3].copy()
latest = world.sort_values('date').groupby('location').last().reset_index()

fig = px.choropleth(
    latest,
    locations="iso_code",
    color="continent",
    hover_name="location",
    hover_data=["total_cases","new_cases","total_deaths"],
    title="COVID-19 by Continent"
)
```

```python
fig.show()
```

```python
fig.write_html("covid_continent_map.html")
```
